```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_numbers = np.random.normal(loc=0, scale=1, size=1000)
```

```python
plt.hist(random_numbers, bins=100, color='green', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Normally Distributed Random Numbers")
plt.savefig("normal_distribution_histogram.pdf")
plt.show()
```
