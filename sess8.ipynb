```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,100)

def sinfunction(val):
    return np.sin(val)

def cosfunction(val):
    return np.cos(val)
```

```python
ysin = sinfunction(x)
ycos = cosfunction(x)

fig, axs = plt.subplots(1,2, figsize=(10,4))

# Left panel: sin(x) vs x
axs[0].plot(x, ysin)
axs[0].set_title("sin(x)")
axs[0].set_xlabel("x")
axs[0].set_ylabel("sin(x)")

# Right panel: cos(x) vs x
axs[1].plot(x, ycos)
axs[1].set_title("cos(x)")
axs[1].set_xlabel("x")
axs[1].set_ylabel("cos(x)")

plt.tight_layout()
plt.savefig("sin_cos_multipanel.pdf")  
plt.show()
```
